```python
import numpy as np
import pandas as pd
import pandas_profiling 
import matplotlib.pyplot as plt
from scipy import misc
```

```python
m=pd.read_csv('https://raw.githubusercontent.com/reddyprasade/Data-Sets-For-Machine-Learnig-and-Data-Science/master/DataSets/Advertising.csv')
```

```python
m
```

```python
plt.plot(m)
plt.show()
```

```python
pandas_profiling.ProfileReport(m)
```

```python
x=m['TV']
y=m['Sales']
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.figure(figsize=(16,9))
```

```python
from scipy import stats
slope,intercept,r_value,p_value,std_error=stats.linregress(x,y)
```

```python
print("slope of the equation is:{}".format(slope))
```

```python
print("intercept of the equation is:{}".format(intercept))
```

```python
print("r_square is:{}".format(r_value**2))
```

```python
print("p_value is:{}".format(p_value))
```

```python
print("standerd error is:{}".format(std_error))
```

```python
plt.scatter(x,y,label='original data')
plt.plot(x,slope*x+intercept,'r',label='predicted data')
plt.legend()
```
